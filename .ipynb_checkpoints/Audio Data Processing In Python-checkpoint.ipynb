```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
from glob import glob
import librosa
import librosa.display
import IPython.display as ipd
from itertools import  cycle
sns.set_theme(style="white", palette=None)
color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]
color_cycle = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])
```

## Terms To Know For Audio In Digital Form:

**Frequency (Hz)**<br>
Frequency describes the differences of wave lengths.<br>
We interperate frequency has high and low pitches.<br><br>
**Intensity (db / power)**<br>
Intensity describes the amplitude (height) of the wave.<br>
<br><br>
**Sample Rate**<br>
Sample rate is specific to how the computer reads in the audio file.<br>
Think of it as the "resolution" of the audio.

## Reading in Audio Files
There are many types of audio files: mp3, wav, m4a, flac, ogg

```python
audio_files = glob(r'data/archive/*/*.wav')
```

```python
#Play Audio File
ipd.Audio(audio_files[0])
```

```python
y, sr = librosa.load(audio_files[0])
#sr -> integer value of sample rate
#y -> raw data
print(f'y: {y[:10]}')
print(f'Shape of y: {y.shape}')
print(f'Sample Rate: {sr}')
```

```python
pd.Series(y).plot(figsize=(10,5), lw=1, title='Raw Audio Example', color=color_pal[0])
plt.show()
```

```python
y_trimmed, _ = librosa.effects.trim(y, top_db=40)
pd.Series(y_trimmed).plot(figsize=(10,5), lw=1, title='Raw Audio Trimmed Example', color=color_pal[1])
plt.show()
```

```python
pd.Series(y[30000:30500]).plot(figsize=(10,5), lw=1, title='Raw Audio Zoomed In Example', color=color_pal[2])
plt.show()
```

## Spectrogram

```python
# Applying a Fourier Transform to Look at Frequencies
# Short Time Fourier Transform
D = librosa.stft(y)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
S_db.shape
```

```python
#Plot the transformed audio data
fig, ax = plt.subplots(figsize=(10,5))
img = librosa.display.specshow(S_db, x_axis='time', y_axis='log', ax=ax)
ax.set_title('Spectrogram Example', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```

## Mel (Melodic) Spectrogram

```python
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128*2)
S_db_mel = librosa.amplitude_to_db(S, ref=np.max)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
img = librosa.display.specshow(S_db_mel, x_axis='time', y_axis='log', ax=ax)
ax.set_title('Mel Spectrogram Example', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```
